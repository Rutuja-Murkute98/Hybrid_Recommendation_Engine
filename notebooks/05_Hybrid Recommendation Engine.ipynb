```python
def get_content_recommendations(
        video_id,
        top_n=20
):
    
    try:
        idx = video_indices[video_id]
    except:
        return []

    similarity_scores = list(
        enumerate(
            content_similarity[idx]
        )
    )

    similarity_scores = sorted(
        similarity_scores,
        key=lambda x: x[1],
        reverse=True
    )

    similarity_scores = similarity_scores[1:top_n+1]

    recommendations = []

    for i, score in similarity_scores:

        recommended_video = (
            content_metadata.iloc[i]["video_id"]
        )

        recommendations.append(
            (
                recommended_video,
                score
            )
        )

    return recommendations
```

```python
sample_video = videos_df["video_id"].iloc[0]

get_content_recommendations(sample_video)
```

```python
def hybrid_recommendation(
        user_id,
        seed_video,
        top_n=10
):

    content_recs = get_content_recommendations(
        seed_video,
        top_n=50
    )

    hybrid_scores = []

    for video_id, content_score in content_recs:

        svd_score = svd_model.predict(
            user_id,
            video_id
        ).est

        final_score = (
            0.4 * content_score +
            0.6 * svd_score
        )

        hybrid_scores.append(
            (
                user_id,
                video_id,
                content_score,
                svd_score,
                final_score
            )
        )

    hybrid_scores = sorted(
        hybrid_scores,
        key=lambda x: x[4],
        reverse=True
    )

    return hybrid_scores[:top_n]
```

```python
sample_user = interactions_df["user_id"].iloc[0]

sample_video = interactions_df["video_id"].iloc[0]

recommendations = hybrid_recommendation(
    sample_user,
    sample_video
)

recommendations
```

```python
sample_users = (
    interactions_df["user_id"]
    .unique()[:100]
)

hybrid_results = []

for user_id in sample_users:

    try:

        watched_video = (
            interactions_df[
                interactions_df["user_id"]
                == user_id
            ]["video_id"]
            .iloc[0]
        )

        recommendations = hybrid_recommendation(
            user_id,
            watched_video,
            top_n=10
        )

        hybrid_results.extend(
            recommendations
        )

    except:
        continue

print(
    "Recommendations Generated:",
    len(hybrid_results)
)
```

```python
hybrid_df = pd.DataFrame(

    hybrid_results,

    columns=[
        "user_id",
        "video_id",
        "content_score",
        "svd_score",
        "hybrid_score"
    ]
)

hybrid_df.head()
```
