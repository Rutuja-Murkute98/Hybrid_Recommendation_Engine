```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")

pd.set_option("display.max_columns", None)

print("Libraries Loaded Successfully")
```

```python
DATA_PATH = r"E:\Reel-Recommendation-Engine\data\raw"

USER_FILE = DATA_PATH + r"\user_features_pure.csv"
VIDEO_BASIC_FILE = DATA_PATH + r"\video_features_basic_pure.csv"
VIDEO_STATS_FILE = DATA_PATH + r"\video_features_statistic_pure.csv"

LOG_STD_1 = DATA_PATH + r"\log_standard_4_08_to_4_21_pure.csv"
LOG_STD_2 = DATA_PATH + r"\log_standard_4_22_to_5_08_pure.csv"

LOG_RANDOM = DATA_PATH + r"\log_random_4_22_to_5_08_pure.csv"

print("Paths Created Successfully")
```

```python
import os

print("User File:", os.path.exists(USER_FILE))
print("Video Basic File:", os.path.exists(VIDEO_BASIC_FILE))
print("Video Stats File:", os.path.exists(VIDEO_STATS_FILE))
print("Log Std 1:", os.path.exists(LOG_STD_1))
print("Log Std 2:", os.path.exists(LOG_STD_2))
print("Log Random:", os.path.exists(LOG_RANDOM))
```

```python
users_df = pd.read_csv(USER_FILE)

video_basic_df = pd.read_csv(VIDEO_BASIC_FILE)

video_stats_df = pd.read_csv(VIDEO_STATS_FILE)

log_std1_df = pd.read_csv(LOG_STD_1)

log_std2_df = pd.read_csv(LOG_STD_2)

log_random_df = pd.read_csv(LOG_RANDOM)

print("All Datasets Loaded Successfully")
```

```python
print("="*60)

print("Users Shape:", users_df.shape)

print("Video Basic Shape:", video_basic_df.shape)

print("Video Statistics Shape:", video_stats_df.shape)

print("Log Standard 1 Shape:", log_std1_df.shape)

print("Log Standard 2 Shape:", log_std2_df.shape)

print("Log Random Shape:", log_random_df.shape)
```

```python
datasets = {
    "Users": users_df,
    "Video Basic": video_basic_df,
    "Video Statistics": video_stats_df,
    "Log Standard 1": log_std1_df,
    "Log Standard 2": log_std2_df,
    "Log Random": log_random_df
}

for name, df in datasets.items():

    print("\n" + "="*100)
    print(name)
    print("="*100)

    print(df.columns.tolist())
```

```python
for name, df in datasets.items():

    print("\n" + "="*100)
    print(name)
    print("="*100)

    missing = df.isnull().sum()

    missing = missing[missing > 0]

    print(missing.sort_values(ascending=False).head(20))
```

```python
for name, df in datasets.items():

    print("\n" + "="*100)
    print(name)
    print("="*100)

    print(df.info())
```

```python
train_logs_df = pd.concat(
    [log_std1_df, log_std2_df],
    ignore_index=True
)

print("Training Logs Shape:")
print(train_logs_df.shape)
```

```python
print("Users Duplicates:",
      users_df.duplicated().sum())

print("Video Basic Duplicates:",
      video_basic_df.duplicated().sum())

print("Video Statistics Duplicates:",
      video_stats_df.duplicated().sum())

print("Training Logs Duplicates:",
      train_logs_df.duplicated().sum())
```

```python
user_missing_cols = [
    'onehot_feat4',
    'onehot_feat12',
    'onehot_feat13',
    'onehot_feat14',
    'onehot_feat15',
    'onehot_feat16',
    'onehot_feat17'
]

for col in user_missing_cols:
    users_df[col] = users_df[col].fillna(-1)
```

```python
video_basic_df['video_duration'] = video_basic_df['video_duration'].fillna(
    video_basic_df['video_duration'].median()
)

video_basic_df['music_type'] = video_basic_df['music_type'].fillna(-1)

video_basic_df['tag'] = video_basic_df['tag'].fillna("unknown")
```

```python
print("Users Missing:")
print(users_df.isnull().sum().sum())

print("Video Basic Missing:")
print(video_basic_df.isnull().sum().sum())

print("Video Statistics Missing:")
print(video_stats_df.isnull().sum().sum())

print("Training Logs Missing:")
print(train_logs_df.isnull().sum().sum())
```

```python
train_logs_df["watch_ratio"] = (
    train_logs_df["play_time_ms"] /
    train_logs_df["duration_ms"]
)

train_logs_df["watch_ratio"] = (
    train_logs_df["watch_ratio"]
    .clip(0, 1)
)
```

```python
train_logs_df["engagement_score"] = (

      train_logs_df["is_like"] * 3

    + train_logs_df["is_comment"] * 4

    + train_logs_df["is_follow"] * 5

    + train_logs_df["is_forward"] * 4

    + train_logs_df["is_profile_enter"] * 2

    + train_logs_df["long_view"] * 2

    - train_logs_df["is_hate"] * 5

)
```

```python
train_logs_df["interaction_score"] = (

      train_logs_df["watch_ratio"] * 5

    + train_logs_df["engagement_score"]

)
```

```python
train_logs_df[
    [
        "watch_ratio",
        "engagement_score",
        "interaction_score"
    ]
].describe()
```

```python
print("Before:", train_logs_df.shape)

train_logs_df = train_logs_df.drop_duplicates()

print("After:", train_logs_df.shape)
```

```python
user_activity = train_logs_df.groupby("user_id").agg({

    "video_id":"count",

    "watch_ratio":"mean",

    "engagement_score":"mean"

}).reset_index()
```

```python
user_activity.columns = [

    "user_id",

    "total_interactions",

    "avg_watch_ratio",

    "avg_engagement"

]
```

```python
user_activity["user_activity_score"] = (

      user_activity["total_interactions"] * 0.5

    + user_activity["avg_watch_ratio"] * 50

    + user_activity["avg_engagement"] * 10

)
```

```python
user_activity.head()
```

```python
video_stats_df["video_popularity_score"] = (

      video_stats_df["play_cnt"] * 0.20

    + video_stats_df["like_cnt"] * 0.30

    + video_stats_df["comment_cnt"] * 0.20

    + video_stats_df["share_cnt"] * 0.15

    + video_stats_df["follow_cnt"] * 0.15

)
```

```python
video_stats_df[
    [
        "video_id",
        "video_popularity_score"
    ]
].head()
```

```python
creator_stats = (

    video_basic_df[["video_id","author_id"]]

    .merge(

        video_stats_df[
            [
                "video_id",
                "video_popularity_score"
            ]
        ],

        on="video_id",

        how="left"

    )
)
```

```python
creator_score = (

    creator_stats

    .groupby("author_id")

    ["video_popularity_score"]

    .mean()

    .reset_index()

)
```

```python
creator_score.columns = [

    "author_id",

    "creator_score"

]
```

```python
creator_score.head()
```

```python
recommendation_dataset = train_logs_df.merge(

    users_df,

    on="user_id",

    how="left"

)
```

```python
print(recommendation_dataset.shape)
```

```python
recommendation_dataset = recommendation_dataset.merge(

    video_basic_df,

    on="video_id",

    how="left"

)
```

```python
print(recommendation_dataset.shape)
```

```python
recommendation_dataset = recommendation_dataset.merge(

    video_stats_df,

    on="video_id",

    how="left"

)
```

```python
print(recommendation_dataset.shape)
```

```python
recommendation_dataset = recommendation_dataset.merge(

    user_activity[
        [
            "user_id",
            "user_activity_score"
        ]
    ],

    on="user_id",

    how="left"

)
```

```python
recommendation_dataset = recommendation_dataset.merge(

    creator_score,

    on="author_id",

    how="left"

)
```

```python
print("Final Dataset Shape:")

print(recommendation_dataset.shape)
```

```python
recommendation_dataset["recommendation_score"] = (

      recommendation_dataset["watch_ratio"] * 0.30

    + recommendation_dataset["engagement_score"] * 0.25

    + recommendation_dataset["user_activity_score"] * 0.15

    + recommendation_dataset["video_popularity_score"] * 0.15

    + recommendation_dataset["creator_score"] * 0.15

)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

recommendation_dataset["recommendation_score"] = scaler.fit_transform(
    recommendation_dataset[["recommendation_score"]]
)
```

```python
recommendation_dataset["recommendation_score"].describe()
```

```python
user_features = recommendation_dataset.groupby("user_id").agg({

    "watch_ratio":"mean",

    "engagement_score":"mean",

    "recommendation_score":"mean",

    "user_activity_score":"mean"

}).reset_index()
```

```python
video_features = recommendation_dataset.groupby("video_id").agg({

    "watch_ratio":"mean",

    "engagement_score":"mean",

    "recommendation_score":"mean",

    "video_popularity_score":"mean"

}).reset_index()
```
